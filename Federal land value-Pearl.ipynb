```python
import json
import pandas as pd
import geopandas as gpd


df = pd.read_excel('/Users/pearlliu100/Documents/GitHub/UDS-Project--Full-Team-/Data/Land-Prices_DLOS_2020_26Oct.xlsx')
df = df.rename(columns=df.iloc[0])
df = df.drop(index=0)
df = df.reset_index()
df = df.drop(columns='index')
df
```

```python
df_la = df[df['County'] == 'Los Angeles County']
df_la
```

```python
df_la_2012 = df_la[df_la['Year'] == 2012]
df_la_2012 = df_la_2012.rename(columns={'Land Value\n(1/4 Acre Lot, Standardized)': 'Land Value\n(1/4 Acre Lot, Standardized)-2012',
                                        'Land Value\n(Per Acre, As-Is)':'Land Value\n(Per Acre, As-Is)-2012',
                                        'Land Share of Property Value':'Land Share of Property Value-2012',
                                        'Property Value (Standardized)':'Property Value (Standardized)-2012',
                                        'Property Value (As-is)':'Property Value (As-is)-2012'})
df_la_2012.set_index('Census Tract',inplace=True)
df_la_2012 = df_la_2012.drop(columns='Year')
df_la_2012.info()
```

```python
df_la_2019  = df_la[df_la['Year'] == 2019]
df_la_2019 = df_la_2019.rename(columns={'Land Value\n(1/4 Acre Lot, Standardized)': 'Land Value\n(1/4 Acre Lot, Standardized)-2019',
                                        'Land Value\n(Per Acre, As-Is)':'Land Value\n(Per Acre, As-Is)-2019',
                                        'Land Share of Property Value':'Land Share of Property Value-2019',
                                        'Property Value (Standardized)':'Property Value (Standardized)-2019',
                                        'Property Value (As-is)':'Property Value (As-is)-2019'})
df_la_2019.set_index('Census Tract',inplace=True)
df_la_2019 = df_la_2019.drop(columns='State')
df_la_2019 = df_la_2019.drop(columns='County')
df_la_2019 = df_la_2019.drop(columns='Year')
df_la_2019.info()
```

```python
joindf = df_la_2012.join(df_la_2019,how='inner')
joindf
```
