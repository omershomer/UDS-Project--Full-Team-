```python
import requests
import json


url = 'https://data.lacounty.gov/resource/rjze-ndj2.json?$where=within_box(the_geom, 34.1164, -118.3217, 34.01333, -118.2675)&$limit=100000' # copied and pasted from the webpage
r = requests.get(url)
parcel = pd.DataFrame(json.loads(r.text))
parcel
```

```python
parcelgdf = gpd.GeoDataFrame(
    parcel, geometry=gpd.points_from_xy(parcel.lon_coord, parcel.lat_coord, 
                                          crs='EPSG:4326'))
parcelgdf
```

```python
gdf_2020_sold = gdf_2020_sold.drop(columns=['index_right'])
gdf_2020_sold
```

```python
joinedgdf2020=gpd.sjoin(gdf_2020_sold.to_crs('EPSG:4326'),parcelgdf.to_crs('EPSG:4326'),how='left')
joinedgdf2020
```

```python
join2020 = gdf_2020_sold.merge(parcelgdf, how='left',on='ain')
join2020 = join2020.dropna()
join2020
```

```python
join2020.query('neighborhood == "Westlake"').neighborhood.count()
```

```python
join2020['shape_area'] = pd.to_numeric(join2020['shape_area'])
join2020['roll_landvalue'] = pd.to_numeric(join2020['roll_landvalue'])
```

```python
join2020['landvalue_sqft']=join2020['roll_landvalue']/join2020['shape_area']
join2020
```

```python
censusjoin2020 = join2020.groupby('CT20').ain.count()
censusjoin2020
```

```python
df_census_n = pd.DataFrame(censusjoin2020)
df_census_n =df_census_n.rename(columns={'ain': 'n_parcel'})
df_census_n
```

```python
censuslvsfjoin2020 = join2020.groupby('CT20').landvalue_sqft.sum()
censuslvsfjoin2020
```

```python
df_lvsf = pd.DataFrame(censuslvsfjoin2020)
df_lvsf
```

```python
df_lvsf.info()
```

```python
df_census_n.info()
```

```python
newjoin = df_lvsf.merge(df_census_n, how='left',on='CT20')
newjoin
```

```python
newjoin['avgprice_tract']=newjoin['landvalue_sqft']/newjoin['n_parcel']
newjoin
```
